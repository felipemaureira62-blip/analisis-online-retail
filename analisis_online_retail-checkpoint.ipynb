# üìä An√°lisis Exploratorio - Online Retail Dataset
## Dataset de transacciones de una tienda online del Reino Unido (2009-2011)

**Objetivo:** Explorar el comportamiento de ventas, productos, pa√≠ses y clientes mediante t√©cnicas de EDA y An√°lisis RFM.

**Herramientas:** Python, Pandas, Matplotlib, Seaborn

```python
# Importar libreruas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("‚úÖ Librer√≠as cargadas correctamente")
```

```python
# Cargar y combinar las dos hojas del archivo
hojas = pd.read_excel(r"C:\Users\felip\Desktop\Python\Plan de estudios\online_retail.xlsx", sheet_name=["2009-2010", "2010-2011"])
df = pd.concat(hojas.values(), ignore_index=True)
print(f"‚úÖ Datos cargados correctamente")
print(f"üìã Filas: {df.shape[0]:,} | Columnas: {df.shape[1]}")
df.head(10)
```

## üßπ Limpieza de Datos

```python
# Ingenier√≠a de features - Extraccion de a√±o, fecha y hora
df["Total_price"] = df["Quantity"] * df["Price"]
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["Fecha"] = df["InvoiceDate"].dt.date
df["A√±o"] = df["InvoiceDate"].dt.year
df["Mes"] = df["InvoiceDate"].dt.month
df["Dia_semana"] = pd.to_datetime(df["Fecha"]).dt.day_name()
df["Hora_num"] = df["InvoiceDate"].dt.hour

# Limpieza - Eliminacion de duplicados
df = df.drop_duplicates()
df["Description"] = df["Description"].fillna("Unknown").replace("?", "Unknown")
df["Customer_ID"] = df["Customer_ID"].fillna("Unknown")

# Segmentaci√≥n - Separacion de valores positivos y negativos
df_filtrado = df[(df["Quantity"] > 0) & (df["Quantity"] < 1000)]
devoluciones = df[df["Quantity"] < 0]

print(f"‚úÖ Limpieza completada")
print(f"üìã Dataset completo: {df.shape[0]:,} filas")
print(f"üìã Dataset filtrado: {df_filtrado.shape[0]:,} filas")
print(f"üìã Devoluciones: {devoluciones.shape[0]:,} filas")
```

## üìà Resumen Estad√≠stico

```python
# Definir dataset
datasets = [("COMPLETO", df), ("FILTRADO", df_filtrado), ("DEVOLUCIONES", devoluciones)]

for nombre, data in datasets:
    print(f"\n{'='*45}")
    print(f"  {nombre}")
    print(f"{'='*45}")
    display(data[["Quantity", "Price", "Total_price"]].describe().T.round(2))
```

## üèÜ Top Ventas

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top 10 productos m√°s vendidos por cantidad
top_productos = df_filtrado.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
axes[0].barh(top_productos.index, top_productos.values, color="steelblue")
axes[0].set_title("Top 10 Productos m√°s Vendidos")
axes[0].set_xlabel("Cantidad total")
axes[0].invert_yaxis()

# Top 10 pa√≠ses por ingresos
top_paises = df_filtrado.groupby("Country")["Total_price"].sum().sort_values(ascending=False).head(10)
axes[1].barh(top_paises.index, top_paises.values, color="coral")
axes[1].set_title("Top 10 Pa√≠ses por Ingresos")
axes[1].set_xlabel("Ingresos totales")
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

## üìÖ An√°lisis Temporal

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Ventas diarias
ventas_diarias = df_filtrado.groupby("Fecha")["Total_price"].sum()
axes[0,0].plot(ventas_diarias.index, ventas_diarias.values, color="steelblue")
axes[0,0].set_title("Ventas Diarias")
axes[0,0].set_xlabel("Fecha")
axes[0,0].set_ylabel("Ingresos")

# Ventas por dia de semana
ventas_semana = df_filtrado.groupby("Dia_semana")["Total_price"].sum()
orden = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
ventas_semana = ventas_semana.reindex(orden)
axes[0,1].bar(ventas_semana.index, ventas_semana.values, color="coral")
axes[0,1].set_title("Ingresos por D√≠a de la Semana")
axes[0,1].set_xlabel("D√≠a")
axes[0,1].set_ylabel("Ingresos")

# Ventas mensuales por a√±o
ventas_mensuales = df_filtrado.groupby(["Mes","A√±o"])["Total_price"].sum().reset_index()
for a√±o in sorted(ventas_mensuales["A√±o"].unique()):
    datos = ventas_mensuales[ventas_mensuales["A√±o"] == a√±o]
    axes[1,0].plot(datos["Mes"], datos["Total_price"], marker="o", label=str(a√±o))
axes[1,0].set_title("Ventas Mensuales por A√±o")
axes[1,0].set_xlabel("Mes")
axes[1,0].set_ylabel("Ingresos")
axes[1,0].legend(title="A√±o")
axes[1,0].grid(True)

# Ventas por hora
axes[1,1].hist(df_filtrado["Hora_num"], weights=df_filtrado["Total_price"], bins=24, color="steelblue")
axes[1,1].set_title("Distribuci√≥n de Ventas por Hora")
axes[1,1].set_xlabel("Hora del d√≠a")
axes[1,1].set_ylabel("Ingresos")

plt.tight_layout()
plt.show()
```

## üë• An√°lisis RFM (Recencia, Frecuencia, Valor Monetario)
> El an√°lisis RFM permite identificar los clientes m√°s valiosos bas√°ndose en su comportamiento de compra.

```python
# Calcular RFM
fecha_referencia = df_filtrado["InvoiceDate"].max() + pd.Timedelta(days=1)

df_rfm = (df_filtrado.groupby("Customer_ID")
          .agg({"InvoiceDate": ["count", "max"], "Total_price": "sum"})
          .reset_index())
df_rfm = df_rfm[df_rfm["Customer_ID"] != "Unknown"]
df_rfm.columns = ["Customer_ID", "Frecuencia", "Ultima_Compra", "Valor_Monetario"]
df_rfm["Recencia"] = (fecha_referencia - df_rfm["Ultima_Compra"]).dt.days

print(f"‚úÖ RFM calculado para {df_rfm.shape[0]:,} clientes")
display(df_rfm.head(10))
print("\nEstad√≠sticas RFM:")
display(df_rfm[["Recencia", "Frecuencia", "Valor_Monetario"]].describe().T.round(2))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Recencia
sns.histplot(df_rfm["Recencia"], bins=50, kde=True, ax=axes[0], color="steelblue")
axes[0].set_title("Distribuci√≥n de Recencia\n(D√≠as desde √∫ltima compra)")
axes[0].set_xlabel("D√≠as")

# Frecuencia
sns.histplot(df_rfm["Frecuencia"], bins=50, kde=True, ax=axes[1], color="coral")
axes[1].set_title("Distribuci√≥n de Frecuencia\n(N√∫mero de transacciones)")
axes[1].set_xlabel("Transacciones")

# Valor Monetario
sns.histplot(df_rfm["Valor_Monetario"], bins=50, kde=True, ax=axes[2], color="green")
axes[2].set_title("Distribuci√≥n de Valor Monetario\n(Ingresos por cliente)")
axes[2].set_xlabel("Ingresos")

plt.tight_layout()
plt.show()
```

## üìâ Curva de Pareto - Concentraci√≥n de Ingresos
> ¬øCu√°ntos clientes generan el 80% de los ingresos?

```python
ingresos_clientes = df_rfm["Valor_Monetario"].sort_values(ascending=False)
porcentaje = ingresos_clientes.cumsum() / ingresos_clientes.sum()
clientes_80 = (porcentaje <= 0.8).sum()

plt.figure(figsize=(10, 6))
porcentaje.reset_index(drop=True).plot(color="steelblue")
plt.axhline(0.8, color="red", linestyle="--", label="80% de ingresos")
plt.axvline(clientes_80, color="orange", linestyle="--", label=f"{clientes_80} clientes")
plt.text(clientes_80 + 50, 0.65, f"{clientes_80} clientes\ngeneran el 80%\nde los ingresos", 
         color="orange", fontsize=11)
plt.title("Curva de Pareto - Concentraci√≥n de Ingresos por Cliente")
plt.xlabel("Clientes ordenados por ingresos")
plt.ylabel("Porcentaje acumulado")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
